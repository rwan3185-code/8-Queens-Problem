```python
board = {
    1:' ', 2:' ', 3:' ',
    4:' ', 5:' ', 6:' ',
    7:' ', 8:' ', 9:' '
}

def printBoard(board):
    print(board[1] + '|' + board[2] + '|' + board[3])
    print('-+-+-')
    print(board[4] + '|' + board[5] + '|' + board[6])
    print('-+-+-')
    print(board[7] + '|' + board[8] + '|' + board[9])
```

```python
def spaceIsFree(position):
    if board[position] == ' ':
        return True
    else:
        return False


def chkDraw():
    for key in board.keys():
        if board[key] == ' ':
            return False
    return True


def chkForWin():
    win_conditions = [
        (1,2,3), (4,5,6), (7,8,9),   # rows
        (1,4,7), (2,5,8), (3,6,9),   # columns
        (1,5,9), (3,5,7)             # diagonals
    ]

    for condition in win_conditions:
        if board[condition[0]] == board[condition[1]] == board[condition[2]] and board[condition[0]] != ' ':
            return True

    return False


def chkMarkForWin(mark):
    win_conditions = [
        (1,2,3), (4,5,6), (7,8,9),
        (1,4,7), (2,5,8), (3,6,9),
        (1,5,9), (3,5,7)
    ]

    for condition in win_conditions:
        if board[condition[0]] == board[condition[1]] == board[condition[2]] == mark:
            return True

    return False
```

```python
player = 'X'
computer = 'O'


def insertLetter(letter, position):
    if spaceIsFree(position):
        board[position] = letter
        printBoard(board)
        print()

        if chkForWin():
            if letter == computer:
                print("Computer wins!")
            else:
                print("Player wins!")
            exit()

        if chkDraw():
            print("Draw!")
            exit()
    else:
        print("Position taken, try again.")
        position = int(input("Enter new position: "))
        insertLetter(letter, position)


def playerMove():
    position = int(input("Enter position for X (1-9): "))
    insertLetter(player, position)


def compMove():
    bestScore = -1000
    bestMove = 0

    for key in board.keys():
        if board[key] == ' ':
            board[key] = computer
            score = minimax(board, False)
            board[key] = ' '

            if score > bestScore:
                bestScore = score
                bestMove = key

    insertLetter(computer, bestMove)


def minimax(board, isMaximizing):

    if chkMarkForWin(computer):
        return 1

    if chkMarkForWin(player):
        return -1

    if chkDraw():
        return 0

    if isMaximizing:
        bestScore = -1000

        for key in board.keys():
            if board[key] == ' ':
                board[key] = computer
                score = minimax(board, False)
                board[key] = ' '
                bestScore = max(score, bestScore)

        return bestScore

    else:
        bestScore = 1000

        for key in board.keys():
            if board[key] == ' ':
                board[key] = player
                score = minimax(board, True)
                board[key] = ' '
                bestScore = min(score, bestScore)

        return bestScore
```

```python
#### printBoard(board)
print("Computer goes first...\n")

while not chkForWin():
    compMove()
    
    if chkForWin() or chkDraw():
        break
        
    playerMove()
```
